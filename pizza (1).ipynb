```python
class Pizza:
    def __init__(self, description, cost):
        self._description = description
        self._cost = cost
    
    def get_description(self):
        return self._description
    
    def get_cost(self):
        return self._cost
#Menünün güzel gözükmesi için burada kullanılır, yoksa sipariş edilemiyor.
class additional(Pizza):
    def __init__(self):
        super().__init__("Additional",0)

class ClassicPizza(Pizza):
    def __init__(self):
        super().__init__("Classic Pizza: Tomato sauce, mozzarella cheese, and pepperoni", 10.99)


class MargheritaPizza(Pizza):
    def __init__(self):
        super().__init__("Margherita Pizza: Tomato sauce, fresh mozzarella, and basil", 8.99)


class TurkPizza(Pizza):
    def __init__(self):
        super().__init__("Turk Pizza: Sucuk, tomatoes, green peppers, and onions", 12.99)


class DominosPizza(Pizza):
    def __init__(self):
        super().__init__("Dominos Pizza: Tomato sauce, cheese, and a choice of toppings", 15.99)
```

```python
class Decorator(Pizza):
    def __init__(self, component):
        self.component = component
    
    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)
    
    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)


class Olives(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self._description = "+Olives"
        self._cost = 1.99


class Mushrooms(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self._description = "+Mushrooms"
        self._cost = 2.99


class GoatCheese(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self._description = "+Goat Cheese"
        self._cost = 3.99


class Meat(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self._description = "+Meat"
        self._cost = 4.99


class Onions(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self._description = "+Onions"
        self._cost = 1.99


class Corn(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self._description = "+Corn"
        self._cost = 2.99
```

```python
def main():
    print("**** MENÜ ****")
    print(ClassicPizza().get_description(), "-", ClassicPizza().get_cost())
    print(MargheritaPizza().get_description(), "-", MargheritaPizza().get_cost())
    print(TurkPizza().get_description(), "-", TurkPizza().get_cost())
    print(DominosPizza().get_description(), "-", DominosPizza().get_cost())
    print("Soslar:")
    temp = additional()
    print(Olives(temp).get_description(), "-", Olives(temp).get_cost())
    print(Mushrooms(temp).get_description(), "-", Mushrooms(temp).get_cost())
    print(GoatCheese(temp).get_description(), "-", GoatCheese(temp).get_cost())
    print(Meat(temp).get_description(), "-", Meat(temp).get_cost())
    print(Onions(temp).get_description(), "-", Onions(temp).get_cost())
    print(Corn(temp).get_description(), "-", Corn(temp).get_cost())

    # Kullanıcının pizza ve sos seçmesini iste
    selected_pizza = input("Lütfen bir pizza seçin (ClassicPizza/MargheritaPizza/TurkPizza/DominosPizza): ")
    selected_sauce = input("Lütfen bir sos seçin (Olives/Mushrooms/GoatCheese/Meat/Onions/Corn): ")

    # Seçilen pizza ve sos nesnelerini oluştur
    pizza_object = globals()[selected_pizza]()
    sauce_object = globals()[selected_sauce](pizza_object)

    # Seçilen pizza üzerine seçilen sosu ekleyerek fiyatı hesapla
    total_cost = sauce_object.get_cost()

    # Kullanıcı bilgilerini iste
    name = input("Adınız: ")
    id_number = input("TC Kimlik Numaranız: ")
    credit_card_number = input("Kredi Kartı Numaranız: ")
    credit_card_password = input("Kredi Kartı Şifreniz: ")

    # Sipariş bilgilerini ekrana yazdır
    print("**** SİPARİŞ BİLGİLERİ ****")
    print("Sipariş Açıklaması: {}".format(sauce_object.get_description()))
    print("Toplam Tutar: {} TL".format(total_cost))
    now = datetime.now()
    hour = now.hour
    minute = now.minute
    print("Sipariş Saati {}:{}.".format(hour, minute))
    print("Ad: {}".format(name))
    print("TC Kimlik No: {}".format(id_number))
    print("Kredi Kartı No: {}".format(credit_card_number))
    print("Kredi Kartı Şifresi: {}".format(credit_card_password))

    field_names = ['Name','ID','Credit Card Number','Passw','Time',"Pizza Description","Pizza Cost"]
    data = {"Name" : name,"ID" : id_number,"Credit Card Number" : credit_card_number,"Passw" :credit_card_password,"Time" :now,"Pizza Description" :sauce_object.get_description(),"Pizza Cost" :sauce_object.get_cost()}

    with open('/content/drive/MyDrive/Colab Notebooks/data.csv', mode='a', newline='') as file:
        dict_object = csv.DictWriter(file, fieldnames=field_names) 
  
        dict_object.writerow(data)
```
